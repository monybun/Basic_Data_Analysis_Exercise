#### Random Normal Distribution

```python
randnorm = np.random.randn(8,8)
randnorm
```

```python
# Two-by-four array of samples from N(3, 6.25):
#np.random.normal(mu, sigma, 1000)
bish = np.random.normal(3, 2.5, size=(2, 4))
bish
```

```python
# Plot Histogram and N-distribution

mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

count, bins, ignored = pp.hist(s, 30, density=True)
pp.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
pp.show()
```
